rank.entropy filter on phase contrast segmentation

```python
from skimage import io,filters,img_as_float,color,morphology
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = io.imread('/content/drive/MyDrive/Colab images/Scratch_Assay.jpg')
plt.imshow(img)
```

```python
img_g = color.rgb2gray(img)
```

```python
otsu_t = img_g>filters.threshold_otsu(img_g)
plt.imshow(otsu_t)
```

```python
entropy = filters.rank.entropy(img_g,morphology.disk(10))
img_e = entropy < filters.threshold_otsu(entropy)
plt.imshow(img_e)
img_seg = morphology.remove_small_objects(img_e,20)
plt.imshow(img_seg)
```

```python
img_seg = morphology.remove_small_objects(img_e,1500)
plt.imshow(img_seg)
```

```python
plt.subplot(221)
plt.imshow(img_g,cmap='gray')
plt.subplot(222)
plt.imshow(otsu_t)
plt.subplot(223)
plt.imshow(img_e)
plt.subplot(224)
plt.imshow(img_seg)
```

```python
area = np.sum(img_seg)
print(area)
```
