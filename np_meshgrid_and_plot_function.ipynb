```python
from skimage import morphology, data,filters,feature,measure
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
a = 2
b = 4
size =250

x, y = np.meshgrid(np.arange(-size//2+1, size//2+1), np.arange(-size//2+1, size//2+1))
kernel1 =np.sin(x)
kernel2 =np.sin(x) + np.cos(y)
kernel3 = x**2/a**2 + y**2/b**2
kernel3_thresh = np.logical_and(kernel3>8,kernel3<10)
#kernel = kernel / np.sum(np.abs(kernel))

#a = np.logical_and(kernel<10,kernel>8)

"""
plt.subplot(141),plt.imshow(kernel1,cmap='gray')
plt.subplot(142),plt.imshow(kernel2,cmap='gray')
plt.subplot(143),plt.imshow(kernel3,cmap='gray')
plt.subplot(144),plt.imshow(kernel3_thresh,cmap='gray')
#plt.tight_layout()

# https://numpy.org/doc/stable/reference/generated/numpy.logical_and.html

# figure too small
"""


fig,ax = plt.subplots(2,2,figsize=(10,10))

ax[0,0].imshow(kernel1,cmap='gray')
ax[0,1].imshow(kernel2,cmap='gray')
ax[1,0].imshow(kernel3,cmap='gray')
ax[1,1].imshow(kernel3_thresh,cmap='gray')

```
