```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import jaccard_score, accuracy_score, precision_score, recall_score
```

```python
def generate_biomodal_2d_data():

    rs = np.random.RandomState(seed=0)

    x1 = rs.normal(3, 1, (150,2))
    x2 = rs.normal(8, 1.5, (150,2))

    x_all = np.concatenate((x1, x2), axis=0)
    rs.shuffle(x_all)
    return x_all
```

```python
def generate_biomodal_2d_data():

    rs = np.random.RandomState(seed=0)

    x1 = rs.normal(3, 1, (150,2))
    x2 = rs.normal(8, 1.5, (150,2))  #wider distribution
    x3 = rs.normal(5, 1, (150,2))
    x4  = rs.normal(6, 1, (150,2))

    x_all = np.concatenate((x1, x2, x3, x4), axis=0)
    rs.shuffle(x_all)
    return x_all
```

```python
data = generate_biomodal_2d_data()

plt.scatter(data[:, 0], data[:, 1], c='#DDDDDD')
plt.xlabel('progression')
plt.ylabel('mutational signature')
```

```python
data = generate_biomodal_2d_data()

plt.scatter(data[:, 0], data[:, 1], c='#DDDDDD')
plt.xlabel('progression')
plt.ylabel('mutational signature')
```

```python
train_data = data[:200]
validation_data = data[200:250]
remaining_data = data[250:]

model = KMeans(n_clusters=2, random_state=0,  n_init="auto").fit(train_data)
```

```python
train_data = data[:400]
validation_data = data[400:550]
remaining_data = data[550:]

model = KMeans(n_clusters=4, random_state=0,  n_init="auto").fit(train_data)
```

```python
result = model.predict(train_data)

colors = ['orange', 'cyan','red','blue']
predicted_colors = []
for i in result:
  predicted_colors.append(colors[i-1])

plt.scatter(train_data[:, 0], train_data[:, 1], c=predicted_colors)
plt.xlabel('progression')
plt.ylabel('mutational signature')

centroids = model.cluster_centers_
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    marker="x",
    s=169,
    linewidths=3,
    color="black",
    zorder=10,
)
```

```python
result = model.predict(validation_data)

colors = ['orange', 'blue','red','blue']
predicted_colors = []
for i in result:
  predicted_colors.append(colors[i-1])

plt.scatter(validation_data[:, 0], validation_data[:, 1], c=predicted_colors)
plt.xlabel('progression')
plt.ylabel('mutational signature')
```

```python
prediction = model.predict(remaining_data)

predicted_colors = [colors[i-1] for i in prediction]

plt.scatter(remaining_data[:, 0], remaining_data[:, 1], c=predicted_colors)
plt.xlabel('progression')
plt.ylabel('mutational signature')
```
