```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# 1D signal
step_signal = np.zeros(100)
step_signal[50:]=1

fig,ax=plt.subplots()
ax.plot(step_signal)
plt.show()
```

```python
np.random.seed(seed=31)
noise = np.random.normal(0,0.35,step_signal.shape)
noisy = step_signal+noise

fig,ax=plt.subplots()
ax.plot(noisy)
plt.show()
```

```python
# local filtering

smooth_1 = (noisy[1:]+noisy[:-1])/2
# or (noisy[:-1] + noisy[1:] + noisy[2:])/3

fig,ax=plt.subplots()
ax.plot(smooth_1)
plt.show()

print(smooth_1.shape)
```

```python
# or (noisy[:-1] + noisy[1:] + noisy[2:])/3
# what if 11 rolling average, becomes curbumsome
# technical difference between convolve and correlate

mean_kernel3=np.full((3,),1/3) # np.full((13,),1/13), 11 rolling average is as easy as 3
print(mean_kernel3)

# he used np.convolve and Exercise 1 used scipy.ndimage.convolve for 1D signal
from scipy import signal
smooth_kernel3 = signal.convolve(noisy,mean_kernel3,mode='valid')
# mode ='same', padded with zero, the edge has a sharp drop off
# 'valid', no padding

fig,ax=plt.subplots()
ax.plot(smooth_kernel3)
plt.show()
```
