```python
rename_map = {
    # ===== Nerve terminal (green / gr) =====
    "cc_distgr": "Cluster_dist_nerve",
    "cc_size_2gr": "Cluster_size_nerve",
    "ccgr": "Cluster_numbers_nerve",
    "fragmentationgr": "Fragmentation_nerve",
    "compactness_1gr": "Non-compactness_nerve",
    "shape_factor_1gr": "Shape factor_nerve",
    "surface_2gr": "Rugosity_int_nerve",
    "surface_3gr": "Rugosity_ext_nerve",
    "max_extentgr": "Length_nerve",
    "surface volume ratio_1gr": "Surf./vol. ratio_nerve",
    "surface_1gr": "Surface_nerve",
    "volume_1gr": "Volume_nerve",

    # ===== Motor endplate (red) =====
    "cc_distred": "Cluster_dist_endplate",
    "cc_size_2red": "Cluster_size_endplate",
    "ccred": "Cluster_numbers_endplate",
    "fragmentationred": "Fragmentation_endplate",
    "compactness_1red": "Non-compactness_endplate",
    "shape_factor_1red": "Shape factor_endplate",
    "surface_2red": "Rugosity_int_endplate",
    "surface_3red": "Rugosity_ext_endplate",
    "max_extentred": "Length_endplate",
    "surface volume ratio_1red": "Surf./vol. ratio_endplate",
    "surface_1red": "Surface_endplate",
    "volume_1red": "Volume_endplate",

# five columns
    "coverage": "Coverage",
    "n_intersectiongr": "Intersection",
    "ave_dist": "Average dist.",
    "com_distgr": "Mass-dist.",
    "haus_dist": "Hausdorff dist.",

    "coverage.1": "Coverage_grmin",
    "n_intersectiongrmin": "Intersection_grmin",
    "ave_dist.1": "Average dist._grmin",
    "com_distgrmin": "Mass-dist._grmin",
    "haus_dist.1": "Hausdorff dist._grmin",

     "haus_dist95": "Hausdorff dist.95",
    "haus_dist95.1": "Hausdorff dist.95_grmin",
}
```

```python
figure2c_order = [
    # nerve terminal
    "Cluster_dist_nerve",
    "Cluster_size_nerve",
    "Cluster_numbers_nerve",
    "Fragmentation_nerve",
    "Non-compactness_nerve",
    "Shape factor_nerve",
    "Rugosity_int_nerve",
    "Rugosity_ext_nerve",
    "Length_nerve",
    "Surf./vol. ratio_nerve",
    "Surface_nerve",
    "Volume_nerve",

    # motor endplate
    "Cluster_dist_endplate",
    "Cluster_size_endplate",
    "Cluster_numbers_endplate",
    "Fragmentation_endplate",
    "Non-compactness_endplate",
    "Shape factor_endplate",
    "Rugosity_int_endplate",
    "Rugosity_ext_endplate",
    "Length_endplate",
    "Surf./vol. ratio_endplate",
    "Surface_endplate",
    "Volume_endplate",

    # interaction
    "Coverage",
    "Intersection",
    "Average dist.",
    "Mass-dist.",
    "Hausdorff dist.",
]
```

```python
from pathlib import Path
import pandas as pd

input_dir = Path(r"/content/drive/MyDrive/Xiawei_nmJ_2ndCombine/LUMBRICAL")
output_dir = Path(r"/content/drive/MyDrive/Xiawei_nmJ_2ndCombine/LUMBRICAL_output")
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
def process_one_csv(csv_path: Path):
    df = pd.read_csv(csv_path)
    df = df.rename(columns=rename_map)

    ordered_existing = [c for c in figure2c_order if c in df.columns]
    remaining_cols = [c for c in df.columns if c not in ordered_existing]
    final_cols = ordered_existing + remaining_cols
    df = df[final_cols]

    out_csv = output_dir / f"{csv_path.stem}_renamed_reordered.csv"
    df.to_csv(out_csv, index=False)
```

```python
def batch_process():
    csv_files = list(input_dir.glob("*.csv"))
    if not csv_files:
        print(f"No CSV files found in: {input_dir}")
        return

    for csv_file in csv_files:
        process_one_csv(csv_file)

    print("\nDone.")

if __name__ == "__main__":
    batch_process()
```
